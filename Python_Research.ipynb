## 1. Variables and Basic Data Types

Variables are used to store data values. Python has several built-in data types, including integers, floats, and strings.

```python
# Integer variable
my_integer = 10
print(f"Integer: {my_integer}, Type: {type(my_integer)}")

# Float variable
my_float = 20.5
print(f"Float: {my_float}, Type: {type(my_float)}")

# String variable
my_string = "Hello, Python!"
print(f"String: {my_string}, Type: {type(my_string)}")

# Boolean variable
my_boolean = True
print(f"Boolean: {my_boolean}, Type: {type(my_boolean)}")
```

## 2. Collection Data Types

Python offers powerful built-in collection types to store multiple items.

```python
# List: Ordered, changeable, allows duplicate members.
my_list = [1, 2, 3, "apple", True]
print(f"List: {my_list}, Type: {type(my_list)}")
print(f"First element of list: {my_list[0]}")
my_list.append(4)
print(f"List after append: {my_list}")

# Tuple: Ordered, unchangeable, allows duplicate members.
my_tuple = (1, 2, 3, "banana")
print(f"Tuple: {my_tuple}, Type: {type(my_tuple)}")
print(f"First element of tuple: {my_tuple[0]}")

# Dictionary: Unordered, changeable, no duplicate keys. Stores key-value pairs.
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
print(f"Dictionary: {my_dict}, Type: {type(my_dict)}")
print(f"Alice's age: {my_dict['age']}")
my_dict["age"] = 31
print(f"Dictionary after age update: {my_dict}")

# Set: Unordered, unchangeable*, no duplicate members. (*items themselves can be mutable but set operations cannot change the set after creation).
my_set = {1, 2, 3, 2, 1, "orange"}
print(f"Set: {my_set}, Type: {type(my_set)}")
my_set.add(4)
print(f"Set after add: {my_set}")
```

## 3. Control Flow

Control flow statements determine the order in which code is executed. Key constructs are `if/elif/else`, `for` loops, and `while` loops.

```python
# Conditional statements (if/elif/else)
x = 15
if x > 20:
    print("x is greater than 20")
elif x > 10:
    print("x is greater than 10 but not 20")
else:
    print("x is 10 or less")

# For loop: Iterating over a sequence
print("\nFor loop examples:")
for i in range(3): # range(3) generates 0, 1, 2
    print(f"For loop iteration: {i}")

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I like {fruit}")

# While loop: Executes as long as a condition is true
print("\nWhile loop example:")
count = 0
while count < 3:
    print(f"While loop count: {count}")
    count += 1
```

## 4. Functions

Functions are blocks of organized, reusable code that perform a single, related action.

```python
# Defining a simple function with no arguments and no return value
def greet():
    print("Hello from a function!")

greet() # Calling the function

# Function with arguments and a return value
def add_numbers(a, b):
    return a + b

result = add_numbers(5, 3)
print(f"\nResult of adding 5 and 3: {result}")

# Function with default argument values
def say_hello(name="Guest"):
    print(f"Hello, {name}!")

say_hello() # Uses default
say_hello("Bob") # Overrides default
```

## 5. Object-Oriented Programming (OOP) Concepts

OOP is a programming paradigm based on the concept of "objects", which can contain data and code.

```python
# 5.1. Classes and Objects
# A class is a blueprint for creating objects.
# An object is an instance of a class.

class Dog:
    # The __init__ method is the constructor
    def __init__(self, name, age):
        self.name = name # Instance variable/attribute
        self.age = age   # Instance variable/attribute

    # Method (function within a class)
    def bark(self):
        return f"{self.name} says Woof!"

    def get_age(self):
        return self.age

# Creating objects (instances of the Dog class)
my_dog = Dog("Buddy", 5)
her_dog = Dog("Lucy", 2)

print(f"\nMy dog's name is {my_dog.name} and is {my_dog.age} years old.")
print(my_dog.bark())

print(f"Her dog's name is {her_dog.name} and is {her_dog.age} years old.")
print(her_dog.bark())
```

```python
# 5.2. Inheritance
# Allows a class to inherit attributes and methods from another class.

class Animal:
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        return "Some generic sound"

class Cat(Animal):
    def __init__(self, name, species="Cat"):
        super().__init__(species) # Call the parent class's constructor
        self.name = name

    # Overriding parent method
    def make_sound(self):
        return f"{self.name} says Meow!"

    def purr(self):
        return f"{self.name} is purring."

my_cat = Cat("Whiskers")
print(f"\nMy cat is a {my_cat.species} named {my_cat.name}.")
print(my_cat.make_sound())
print(my_cat.purr())
```

```python
# 5.3. Encapsulation (using properties)
# Restricting direct access to some of an object's components.
# In Python, we use conventions (e.g., prefixing with _) or properties.

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance # Private attribute (by convention, not strict enforcement)

    # Getter method
    def get_balance(self):
        return self.__balance

    # Setter method
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. New balance: {self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

my_account = BankAccount(1000)
print(f"\nInitial balance: {my_account.get_balance()}")
my_account.deposit(200)
my_account.withdraw(500)
my_account.withdraw(1000) # Insufficient funds

# Accessing the 'private' attribute directly (discouraged but possible in Python)
# print(my_account.__balance) # This will raise an AttributeError because of name mangling
```

```python
# 5.4. Polymorphism
# Ability of an object to take on many forms. Often seen with method overriding and interfaces.

class Bird:
    def fly(self):
        return "Birds can fly."

class Penguin(Bird):
    def fly(self):
        return "Penguins cannot fly, but they can swim!"

class Sparrow(Bird):
    def fly(self):
        return "Sparrows fly short distances."

# Function that takes a Bird object (or any object with a 'fly' method)
def make_it_fly(bird_obj):
    print(bird_obj.fly())

print("\nPolymorphism example:")
make_it_fly(Bird())
make_it_fly(Penguin())
make_it_fly(Sparrow())
```
